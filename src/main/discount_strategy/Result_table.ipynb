```python
def set_segment_probability(probability_0, probability_1, probability_2):
    segment_probability = {}
    segment_probability[0] = probability_0/100
    segment_probability[1] = probability_1/100
    segment_probability[2] = probability_2/100
    
    return segment_probability
a = %run ./scenarios_and_strategies.ipynb
```

```python
def write_table_strategy_comparison(file_name, exp_data):
    with open(  os.getcwd() + "\\Sources\\" + file_name, mode='a',newline='', encoding="utf8") as file:
        writer = csv.writer(file, delimiter=',')
        row = [exp_data['no_of_customers'],
              exp_data['mu_1'],
              exp_data['mu_2'],
              exp_data['mu_3'],
              exp_data['discount']*100,
              exp_data['%_in_PUP'],
              exp_data['cost_det'],
              exp_data['exp_cost_stoch'],
              exp_data['exp_cost_no_discount'],
              exp_data['gap_no_discount'],
              exp_data['exp_cost_ring-star'],
              exp_data['gap_ring-star']
              ]
        writer.writerow(row)  
        
def write_table_strategy_comparison2(file_name, exp_data):
    with open(  os.getcwd() + "\\Sources\\" + file_name, mode='a',newline='', encoding="utf8") as file:
        writer = csv.writer(file, delimiter=',')
        row = [exp_data['no_of_customers'],
              exp_data['mu_1'],
              exp_data['mu_2'],
              exp_data['mu_3'],
              exp_data['discount']*100,
              exp_data['exp_cost_stoch'],
              exp_data['sample_cost_stoch'],
              exp_data['exp_cost_no_discount'],
              exp_data['sample_cost_no_discount'],
              exp_data['exp_cost_ring-star'],
              exp_data['sample_cost_ring-star'],
              exp_data['sample_size']
              ]
        writer.writerow(row)   

        
def write_table_strategy_comparison3(file_name, exp_data):
    with open(  os.getcwd() + "\\Sources\\" + file_name, mode='a',newline='', encoding="utf8") as file:
        writer = csv.writer(file, delimiter=',')
        row = [exp_data['no_of_customers'],
              exp_data['mu_1'],
              exp_data['mu_2'],
              exp_data['mu_3'],
              exp_data['discount']*100,
              #exp_data['sample_cost_stoch'],
              exp_data['sample_cost_no_discount'],
              exp_data['sample_cost_ring-star'],
              exp_data['sample_size']
              ]
        writer.writerow(row)
        
def write_table_result(file_name, exp_data):
    with open(  os.getcwd() + "\\Sources\\" + file_name, mode='a',newline='', encoding="utf8") as file:
        writer = csv.writer(file, delimiter=',')
        writer.writerow(['Deterministic cost of ring-star','MEAN_cost_no_discount', 'STD_no_discount','MEAN_cost_ring-star','STD_ring_star'])  
        row = [exp_data['cost_det'],
              exp_data['Average_cost_nodiscount'],
              exp_data['STD_nodiscount'],
              exp_data['Average_cost_ring-star'],
              exp_data['STD_ring_star']
              ]
        writer.writerow(row)  
        
```
